```python
import pandas as pd 
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
```

```python
data = "./application_train.csv"

df = pd.read_csv(data)
```

```python
df.head()
```

```python
df.describe()
```

```python
df.info()
```

```python
df1 = df.dropna(axis='columns')
df1.info()
df1.columns.str.startswith('FLAG')
df2 = df1.loc[:,~df1.columns.str.startswith('FLAG')]
df2.info()
```

```python
df3 = df2[["TARGET","NAME_CONTRACT_TYPE","CODE_GENDER","CNT_CHILDREN","AMT_INCOME_TOTAL","AMT_CREDIT", "NAME_INCOME_TYPE","NAME_EDUCATION_TYPE","NAME_FAMILY_STATUS", "NAME_HOUSING_TYPE"]]
df3
```

```python
sns.set(style="whitegrid")

sns.countplot(y = "TARGET", hue= "CODE_GENDER", data = df3, palette="Set1")
```

```python

sns.countplot(y = "TARGET", hue="NAME_CONTRACT_TYPE", data = df3, palette="Set1")
```

```python
sns.countplot(y = "TARGET", hue="NAME_EDUCATION_TYPE", data = df3, palette="Set1")
```

```python
sns.countplot(y = "TARGET", hue="NAME_FAMILY_STATUS", data = df3, palette="Set1")
```

```python
g = sns.catplot(  x="TARGET",y="AMT_CREDIT", hue="CODE_GENDER", data=df3,
                height=6, kind="bar", palette="Set3")
g.despine(left=True)
```

```python
g = sns.catplot(  x="TARGET",y="AMT_INCOME_TOTAL", hue="CODE_GENDER", data=df3,
                height=6, kind="bar", palette="Set3")
g.despine(left=True)
```

```python
sns.jointplot("AMT_INCOME_TOTAL", "AMT_CREDIT", data=df3, xlim=(0, 5000000), ylim=(0, 5000000), kind="reg", color="m")
```

```python
sns.boxplot(x = "TARGET", y = 'AMT_CREDIT',data = df3, palette="Set3")
```

```python
sns.violinplot(x = "CODE_GENDER",y="AMT_CREDIT", hue="TARGET",
                        data= df3, palette="Set3", split=True)
```
